```python
# Get unique values in category and sub-category column
crowdfunding_info_df['category & sub-category'].unique()
```

```python
# Use str.split() to separate cat + sub-cat
crowdfunding_info_df[['category', 'subcategory']] = (
    crowdfunding_info_df['category & sub-category'].str.split('/', n=1, expand=True))

crowdfunding_info_df.head(10)
```

```python
# Get number of unique values in columns
print(crowdfunding_info_df['category'].nunique())
print(crowdfunding_info_df['subcategory'].nunique())
```

```python
# Get unique categories and subcats in lists
categories = crowdfunding_info_df['category'].unique()
subcategories = crowdfunding_info_df['subcategory'].unique()
print(categories)
print(subcategories)
```

```python
# start making numbers to assign for category + sub-cat IDS
import numpy as np
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)
```

```python
# Modify ID numbers
# Add cat0 to each category_id, scat0 for subcategory_ids
cat_ids = ['cat0' + str(cat_id) for cat_id in category_ids]
scat_ids = ['scat0' + str(scat_id) for scat_id in subcategory_ids]

print(cat_ids)
print(scat_ids)
```

```python
# Create new DFs for Cats and Subcats
category_df = pd.DataFrame({
    'category_id': cat_ids,
    'category': categories
})

subcategory_df = pd.DataFrame({
    'subcategory_id': scat_ids,
    'subcategory': subcategories
})
```
